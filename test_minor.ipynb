```python
import os
import re
import glob
import torch
import fitz  # PyMuPDF
from typing import List
from PIL import Image
from llama_index.core import Document, Settings, StorageContext, load_index_from_storage, GPTVectorStoreIndex
from llama_index.embeddings.openai import OpenAIEmbedding
from llama_index.llms.openai import OpenAI
from transformers import BlipProcessor, BlipForConditionalGeneration

# Watermark patterns to remove
WATERMARK_PATTERNS = [
    r"(?i)do not copy",
    r"(?i)sample paper",
    r"(?i)property of .*?",
    r"(?i)for evaluation only",
    r"(?i)FITJEE"
]

def clean_text(text: str) -> str:
    for pat in WATERMARK_PATTERNS:
        text = re.sub(pat, "", text)
    text = re.sub(r"\s{2,}", " ", text)
    return text.strip()

# Load BLIP model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
caption_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

def caption_image(image_path: str) -> str:
    raw_image = Image.open(image_path).convert("RGB")
    inputs = processor(raw_image, return_tensors="pt")
    with torch.no_grad():
        out = caption_model.generate(**inputs)
    return processor.decode(out[0], skip_special_tokens=True)

def extract_text_and_images(pdf_path: str, output_dir: str = "extracted") -> List[Document]:
    os.makedirs(output_dir, exist_ok=True)
    docs = []
    doc = fitz.open(pdf_path)
    base_fn = os.path.splitext(os.path.basename(pdf_path))[0]

    for page_idx, page in enumerate(doc, start=1):
        raw_text = page.get_text()
        text = clean_text(raw_text)
        image_paths = []
        captions = []

        for img_i, img in enumerate(page.get_images(full=True), start=1):
            xref = img[0]
            base_image = doc.extract_image(xref)
            img_bytes = base_image["image"]
            ext = base_image["ext"]
            img_name = f"{base_fn}_page{page_idx}_img{img_i}.{ext}"
            img_path = os.path.join(output_dir, img_name)
            with open(img_path, "wb") as f:
                f.write(img_bytes)
            image_paths.append(img_path)

            try:
                cap = caption_image(img_path)
            except Exception:
                cap = ""
            captions.append(cap)

        content = text
        for idx, cap in enumerate(captions, start=1):
            content += f"\n[IMAGE {idx} caption]: {cap}\n"

        metadata = {
            "source": base_fn,
            "page": page_idx,
            "images": image_paths
        }
        docs.append(Document(text=content, metadata=metadata))

    return docs

# Directories
INPUT_DIR = "./data"
OUTPUT_IMG_DIR = "./extracted"
PERSIST_DIR = "./storage"

# --- SETTINGS: Use OpenAI LLM + Embedding ---
Settings.llm = OpenAI(model="gpt-3.5-turbo", temperature=0.3)
Settings.embed_model = OpenAIEmbedding(model="text-embedding-ada-002")

# --- Indexing ---
if not os.path.exists(PERSIST_DIR):
    all_docs = []
    for pdf_file in glob.glob(os.path.join(INPUT_DIR, "*.pdf")):
        print(f"Parsing \"{pdf_file}\"…")
        docs = extract_text_and_images(pdf_file, output_dir=OUTPUT_IMG_DIR)
        all_docs.extend(docs)

    index = GPTVectorStoreIndex.from_documents(all_docs)
    index.storage_context.persist(persist_dir=PERSIST_DIR)
    print("Index built & saved.")
else:
    storage_ctx = StorageContext.from_defaults(persist_dir=PERSIST_DIR)
    index = load_index_from_storage(storage_ctx)
    print("Index loaded from disk.")

# --- Querying ---
query_engine = index.as_query_engine()
resp = query_engine.query("Generate 50 difficult maths questions on conic sections, include diagrams if relevant and also provide solutions.")
print(resp)
```

```python
import os
from pathlib import Path
import fitz
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration, CLIPProcessor, CLIPModel
from llama_index.core import Document, VectorStoreIndex, Settings
from llama_index.embeddings.openai import OpenAIEmbedding
from llama_index.llms.openai import OpenAI
from sentence_transformers import SentenceTransformer

Settings.llm = OpenAI(model="gpt-3.5-turbo", temperature=0.1)
Settings.embed_model = OpenAIEmbedding(model="text-embedding-3-small")

blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
text_embed_model = SentenceTransformer('all-MiniLM-L6-v2')

def extract_text_and_images(pdf_path: str, output_dir: str):
    os.makedirs(output_dir, exist_ok=True)
    doc = fitz.open(pdf_path)
    data = []
    for i, page in enumerate(doc):
        text = page.get_text() or ""
        images = []
        for img_index, img in enumerate(page.get_images(full=True)):
            xref = img[0]
            base_image = doc.extract_image(xref)
            img_bytes = base_image['image']
            ext = base_image['ext']
            img_path = Path(output_dir) / f"{Path(pdf_path).stem}_page{i+1}_img{img_index+1}.{ext}"
            with open(img_path, 'wb') as f:
                f.write(img_bytes)
            images.append(str(img_path))
        data.append({"source": Path(pdf_path).name, "page": i+1, "text": text, "images": images})
    return data

def get_image_caption(image_path: str) -> str:
    image = Image.open(image_path).convert("RGB")
    inputs = blip_processor(image, return_tensors="pt")
    out = blip_model.generate(**inputs)
    return blip_processor.decode(out[0], skip_special_tokens=True)

def get_image_embedding(image_path: str):
    image = Image.open(image_path).convert("RGB")
    inputs = clip_processor(images=image, return_tensors="pt")
    outputs = clip_model.get_image_features(**inputs)
    return outputs.detach().numpy()[0]

def get_text_embedding(text: str):
    return text_embed_model.encode(text)

def build_documents_from_extracted(extracted):
    docs = []
    for item in extracted:
        src = item['source']
        page_text = item['text'].strip()
        if page_text:
            docs.append(Document(text=page_text, metadata={"source_pdf": src, "page": item['page'], "type": "text"}))
        for img_path in item['images']:
            caption = get_image_caption(img_path)
            docs.append(Document(text=f"[Image from {src} page {item['page']}]: {caption}", metadata={"source_pdf": src, "page": item['page'], "type": "image", "image_path": img_path}))
    return docs

def build_and_query_index(docs, query: str):
    index = VectorStoreIndex.from_documents(docs)
    engine = index.as_query_engine()
    return engine.query(query)

if __name__ == "__main__":
    input_dir = Path("data")
    images_root = Path("output/images")
    all_extracted = []
    for pdf_path in input_dir.glob("*.pdf"):
        out_dir = images_root / pdf_path.stem
        extracted = extract_text_and_images(str(pdf_path), str(out_dir))
        all_extracted.extend(extracted)
    documents = build_documents_from_extracted(all_extracted)
    user_query = "give me 10 questions on chemistry chemical equations"
    response = build_and_query_index(documents, user_query)
    print(response)
```
